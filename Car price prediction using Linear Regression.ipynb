```python
enc_car2 = pd.get_dummies(new_car, columns=['make','fuel-type','aspiration','body-style','drive-wheels',
                                          'engine-location','engine-type','fuel-system'], drop_first=True)
```

```python
plt.figure(figsize=(40,40))
sns.heatmap(enc_car2.corr(), annot=True ,cmap=('summer'))
```

```python
new_car2 = enc_car2.drop(['make_honda','make_peugot','make_porsche','make_subaru',
                         'engine-location_rear','engine-type_l','engine-type_ohcf','engine-type_rotor',
                    'fuel-system_4bbl','fuel-system_idi'], axis=1)
```

```python
norm_car=(new_car2-new_car2.min())/(new_car2.max()-new_car2.min()) # Normalization Step
```

```python
norm_car.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
train,test=train_test_split(norm_car,test_size=0.3)
# dividing the data into train and test
```

```python
len(train)
```

```python
train_x=train.drop('price',axis=1)
train_y=train['price']

test_x=test.drop('price',axis=1)
test_y=test['price']

# we must identify the features from the label
# axis=1 when we drop that means that this is column not row
```

```python
lr=LinearRegression()
```

```python
lr.fit(train_x,train_y)  #training the model
```

```python
pred_y=lr.predict(test_x) # predict on the test set 
```

```python
mean_squared_error(pred_y,test_y) # comparing the prediction with actual values
# method for evalauting the model
```
